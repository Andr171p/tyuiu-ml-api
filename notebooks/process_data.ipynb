```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"ТИУ Бакалавриат Абитуриенты.csv")
df.head(3)
```

```python
df = df.drop("Unnamed: 0", axis=1)
df.head(5)
```

```python
index_names = df[df['Сумма баллов'] == 0].index
df.drop(index_names, inplace=True)
```

```python
def get_avg_score_by_year(direction: str, year: int) -> float:
    filtered_df = df[df["Направление подготовки"] == direction]
    filtered_df = filtered_df[filtered_df["Год приёмной компании"] == year]
    filtered_df = filtered_df[filtered_df["Приказ о зачислении"] == 1]
    # return filterd_df["Сумма баллов"].mean()
    return filtered_df["Сумма баллов"].min()
```

```python
def scores_curve_plot(direction: str) -> None:
    years = list(range(2019, 2024))
    avg_scores = []
    for year in years:
        avg_score = get_avg_score_by_year(
            direction=direction,
            year=year
        )
        avg_scores.append(avg_score)
    fig, ax = plt.subplots(1)
    ax.plot(years, avg_scores, label='Средний балл')
    ax.set_xlabel('Год')
    ax.set_ylabel('Баллы')
    plt.show()
```

```python
scores_curve_plot("08.03.01 Строительство")
```
